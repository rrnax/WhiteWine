Na wykresach rozrzutu nie znalazłem żadnej interesującej rzecy szczególnie, że są dosyć obszerne.\
W cely wykonania grupowania posłuże sie algorytmem DBSCAN, aby go użyć najpierw określe parametry sąsiedztwa i liczby obserwacji. Także podzielimy zbiór na uczący i testowy.\
Grupowanie przeprowadzimy najpierw na testowy a potem na uczącym.

```python
min_observations = 2 * (len(white_wines.columns) - 1)

from sklearn.model_selection import train_test_split
white_wines.reset_index(drop=True, inplace=True)
seed_value = 308200
X = white_wines.drop("Quality", axis=1)
y = white_wines["Quality"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=seed_value)
X_train_groups = X_train.copy()
```

Jako minimalną liczbę obserwacji przyjołem podwojony wymiar przestrzeni danych.

```python
from sklearn.neighbors import NearestNeighbors
neighbors = NearestNeighbors(n_neighbors= min_observations + 1)
neighbors.fit(X_train_groups)
```

```python
dist, ind = neighbors.kneighbors(X_train_groups)
neighbors_distances = sorted(dist[:, -1], reverse=True)
```

```python
plt.figure(figsize=(12,7))
plt.plot(neighbors_distances)
plt.ylabel(f"Distance from {min_observations} nearest neighbour")
plt.xlabel(f"Observation sorted in dependence of {min_observations} nearest neighbors")
plt.show()
```

Dość łatwo tu zlokalizować punkt łokćia stwierdziłbym że swpółrzędna y ów punktu wynosi 20, więc taką wartoś parametu sąsiedztwa przyjmijmy.

```python
from sklearn.cluster import DBSCAN
train_dbscan = DBSCAN(eps=20, min_samples=min_observations)
train_dbscan.fit(X_train_groups)
```

```python
pd.Series(train_dbscan.labels_).value_counts().to_frame()
```

Niestety powstała tylko jedna grupa, więc musimy pobawić się z parametrem 'eps' może uda się uzyskać lepsze wyniki.\
Po wielu próbach najwięcej grup tworzy się prz parametrze eps = 5.9.

```python
train_dbscan = DBSCAN(eps=5.8, min_samples=min_observations)
train_dbscan.fit(X_train_groups)
pd.Series(train_dbscan.labels_).value_counts().to_frame()
```

```python
white_wines_groups = X_train_groups.copy()
white_wines_groups['Group'] = train_dbscan.labels_
outsiders = white_wines_groups[white_wines_groups['Group'] == -1]
fig, (axes) = plt.subplots(6,2,figsize=(20,50))

for i in [0, 1, 2, 3, 4, 5]:
    sns.scatterplot(data = white_wines_groups[white_wines_groups['Group'] != -1], x = white_wines_groups.columns[i], y = white_wines_groups.columns[i+1], hue = 'Group', ax = axes[i][0], palette = 'Set1', s = 45)
    sns.scatterplot(data = white_wines_groups[white_wines_groups['Group'] != -1], x = white_wines_groups.columns[i+2], y = white_wines_groups.columns[i+3], hue = 'Group', ax = axes[i][1], palette = 'Set1', s = 45)
```

Spoglądając na różne wykresy po pogrupowaniu, można zauważyć wyszczególnione wyrażnie grupy.\ 
Grupa 4 jest to grupa o dużej gęstości oraz zawartości wszelkich siarczanów, chlorków i innych składników.\
Grupe 3 bardzo ciężko zdefiniować jest takim przeciętniakiem we wszystkim.\
Grupe 2 wyrużnia sie największą zawartością wolnego dwutlenku siarki i przeciętną jakością.\
Grupa 1 równiez bardzo ciężko zdefiniować.\
Grupa 0 to grupa obszerna z niższą zawartością wszelkich składników niż pozostałe grupy.

```python
train_dbscan.fit(X_test)
pd.Series(train_dbscan.labels_).value_counts().to_frame()
```

Po zastosowaniu chematu grupwania uzyskanego ze zbioru uczącego na zbioorze testowym wynikają tylko 2 grupy.\
