```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option("display.precision", 3)
diagram_colors = ["#fc9521", "#25216b","#7a2a2c", "#00fc6d", "#7b3b7d", "#72A0E0", "#000000"]
```

```python
white_wines.Quality.value_counts().plot(kind="pie", figsize=(10,10), startangle=90, title="Amount of wines in quality classes", label="", colors=diagram_colors)
```

Jak widzimy najmniej jest win bardzo słabych jakościowo i bardzo dobrych jakościowo. Około 75% win ze zbioru pochodzi z klasy 3 i 4 odnośnie jakości.\
Zbadajmy teraz zależność zmiennych od siebie nawzajem.

```python
plt.figure(figsize=(12,12))
sns.heatmap(white_wines.corr(),
            cbar=True,
            annot=True,
            square=True,
            cmap="Blues")
```

Widzimy najsilniejsze korelacje pomiędzy gęstością a cukrem resztkowym lub alkoholem, aałkowitym dwutlenkiem węgal a wolnym dwutlenkiem węgla czego można było się spodziewać.\
Zbadajmy teraz zależność jakości wina od zmiennych.

```python
for variable in white_wines.columns[0:11]:
    plt.figure(figsize=(7,7))
    selected_variables = white_wines[[variable, "Quality"]]
    plt.title("Correlation between variable Quality and " + variable)
    sns.heatmap(selected_variables.corr(),
            cbar=True,
            annot=True,
            square=True,
            cmap="Oranges")
```

Dwie zmienne zawartość kwasu cytrynowego i wolnego dwutlenku węgla mają współczyniki korelacji ze zmienną celu równy odpowiednio 0.0087 i 0.0014.\
Są to wyniki tak bliskie 0, że rezygnuje z użycia ich w budowaniu modelów klasyfikacji i regresji, ponieważ nie wpłyną one w znaczący sposób na wyniki.\
Największy wpływ za to na jakość wina mają zmienne alkochol, gęstość, chlorki.
